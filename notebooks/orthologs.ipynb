```python
orth = '../out/orthologs_correlation.tsv'
dup = '../out/duplicates_correlation.tsv'
```

```python
# plotting imports
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')

plt.rc('font', size=12)
```

```python
# other imports
import itertools
import numpy as np
import pandas as pd
```

```python
r = pd.read_table(orth)
d = pd.read_table(dup)
```

```python
genes = {x for x in d[d['genes'] == 'same']['gene']}
```

Only genes with duplicated KOs
---

```python
plt.figure(figsize=(9, 3))

v = r.set_index('gene')
v = v.loc[v.index.intersection(genes)]

sns.kdeplot(v[v['genes'] == 'same'].groupby('gene').mean()['corr'],
            label='orthologs',
            lw=2.5,
            color='r',
            zorder=10)
# sns.kdeplot(r[r['genes'] == 'shuffled'].groupby('gene').mean()['corr'],
#             label='random gene pairs',
#             color='grey',
#             shade=True,
#             alpha=0.3)
plt.legend(title='gene set',
           frameon=True)
plt.xlabel('average s-score correlation\n(across all strains)')
plt.yticks([])
plt.xlim(-1.05, 1.05)
plt.axvline(0,
            color='grey',
            linestyle='dashed',
            zorder=0)
sns.despine(left=True);
```

```python
plt.figure(figsize=(9, 3))

strains = set(r['strain1']).union(r['strain2'])
for s1, s2 in itertools.combinations(strains, 2):
    sns.kdeplot(v[(v['genes'] == 'same') &
                  (((v['strain1'] == s1) & ((v['strain2'] == s2))) |
                  ((v['strain2'] == s1) & ((v['strain1'] == s2))))
                 ].groupby('gene').mean()['corr'],
                label=s1 + '/' + s2,
                lw=2.5)

plt.legend(title='strains pair',
           frameon=True)
plt.xlabel('average s-score correlation')
plt.yticks([])
plt.xlim(-1.05, 1.05)
plt.axvline(0,
            color='grey',
            linestyle='dashed',
            zorder=0)
sns.despine(left=True);
```

All genes

```python
plt.figure(figsize=(9, 3))

sns.kdeplot(r[r['genes'] == 'same'].groupby('gene').mean()['corr'],
            label='orthologs',
            lw=2.5,
            color='r',
            zorder=10)
sns.kdeplot(r[r['genes'] == 'shuffled'].groupby('gene').mean()['corr'],
            label='random gene pairs',
            color='grey',
            shade=True,
            alpha=0.3)
plt.legend(title='gene set',
           frameon=True)
plt.xlabel('average s-score correlation\n(across all strains)')
plt.yticks([])
plt.xlim(-1.05, 1.05)
plt.axvline(0,
            color='grey',
            linestyle='dashed',
            zorder=0)
sns.despine(left=True);
```
